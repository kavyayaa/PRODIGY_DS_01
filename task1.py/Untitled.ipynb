```python
df = pd.read_csv("data/data.csv", skiprows=4)
```

```python
df_merged = df.merge(country, on="Country Code", how="left")
df_merged.head()
```

```python
df_merged = df_merged.merge(indicator, left_on="Indicator Code", right_on="INDICATOR_CODE", how="left")
df_merged.head()
```

```python
df_merged = df.merge(country, on="Country Code", how="left")
df_merged.head()
```

```python
import matplotlib.pyplot as plt
df_merged["IncomeGroup"].value_counts().plot(kind="bar")
plt.title("Income Group Distribution")
plt.xlabel("Income Group")
plt.ylabel("Count")
plt.savefig("images/income_group.png")
plt.show()
```

```python
df_merged["Region"].value_counts().plot(kind="bar")

plt.title("Region-wise Country Distribution")
plt.xlabel("Region")
plt.ylabel("Count")
plt.savefig("images/region_distribution.png")
plt.show()
```
